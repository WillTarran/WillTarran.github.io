```python
# A second slide might have some code...
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rnd = np.random.rand(2,20)
x = np.linspace(0, 1, 101)
fig, ax = plt.subplots()
ax.plot(x, norm.pdf(x, loc=0.35, scale=0.15), 'k:')
ax.scatter(rnd[0,], 2.5 * rnd[1,], marker='x', c='green')
ax.text(0.65, 2.4, "Fit as a\nButcher's Dog", fontsize=12)
```
