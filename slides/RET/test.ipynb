```python
import numpy as np
import pandas as pd
from scipy import stats
import scipy
print(scipy.__version__)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(20)
wghts = np.random.rand(20)
df = pd.DataFrame(data={'Flat': x, 'Wght': wghts})
under = stats.gaussian_kde(x, bw_method='silverman')
over = stats.gaussian_kde(x, bw_method='silverman', weights=wghts.tolist())
```

```python
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(10,10))
ax1.plot(x, under.pdf(x))
ax2.plot(x, over.pdf(x))
```
