## Review of example electricity market data for illustration of credible price distribution

US wholesale contract data is available from [US Energy Information Administration](https://www.eia.gov/electricity/wholesale/).  Analysis here is of 2018 data on wholesale electricity contracts reported by [Intercontinental Exchange](https://www.theice.com/index) for the trading hubs shown:

<img src="hub-map.png" alt="US trading Hubs" style="width:50%;">

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('US_2018_wholesale.csv', parse_dates=[1,2,3], thousands=',', usecols=range(9))
data['Avg Price $/MWh'].describe()
```

```python
fig, [ax1, ax2] = plt.subplots(2,1, sharex=True, figsize=(10,10), gridspec_kw = {'height_ratios':[3,1]})
sns.kdeplot(data['Avg Price $/MWh'], shade=True, ax=ax1, bw='silverman', legend=False) #note, silverman probably over-smooth
ax1.set_ylabel('Contract Price Distribution', fontsize=14)
sns.boxplot(data['Avg Price $/MWh'], ax=ax2)
ax2.xaxis.label.set_fontsize(14)
fig.subplots_adjust(hspace=0)
```

Graph above represents average contract value, independent of delivered volume.  Calculated Power Volume weighted distribution for supply

```python
data['Contract Days'] = (data['Delivery end date'] - data['Delivery start date']).dt.days + 1
data['Total Delivered'] = data['Daily volume MWh'] * data['Contract Days']
data['Contract Value'] = data['Avg Price $/MWh'] * data['Total Delivered']
data.head()
```

```python
print('Weighted Mean Contract Price:',
       round((data['Avg Price $/MWh'] * data['Total Delivered']).sum() / data['Total Delivered'].sum(), 2))
```

### Use Weighted kde for assessment of contracts adjusted by delivered power

Python plotting libraries do not have builtin weighted kde functions.  Weighted GaussianKDE is available in Scipy Stats v1.2.0+

Temporary upgrade to 1.2.0 works to process data, but I assume will be lost on fresh log in...

```python
from scipy.stats import gaussian_kde
import scipy ; print(scipy.__version__)
```

```python
weighted_avg = gaussian_kde(data['Avg Price $/MWh'], bw_method=0.08, weights=data['Total Delivered'].astype('float64'))
fig, [ax1, ax2] = plt.subplots(2,1, sharex=True, figsize=(10,10), gridspec_kw = {'height_ratios':[3,1]})
x = np.linspace(0,400,4001)
ax1.plot(x, weighted_avg.pdf(x), 'k', label='Weighted by Delivered Power')
sns.kdeplot(data['Avg Price $/MWh'], shade=True, ax=ax1, bw='silverman', legend=True)
sns.boxplot(data['Avg Price $/MWh'], ax=ax2)
ax1.set_ylabel('Relative Distribution', fontsize=14)
ax2.xaxis.label.set_fontsize(14)
fig.subplots_adjust(hspace=0)
```

### estimation of bandwidth for scipy stats matching seaborn 'silverman'

```python
x = np.linspace(0,400,4001)
fig, ax = plt.subplots(figsize=(10,10))
sns.kdeplot(data['Avg Price $/MWh'], shade=True, ax=ax, bw='silverman', legend=True)
for i in np.linspace(0.05, 0.1, 10):
    k = gaussian_kde(data['Avg Price $/MWh'], bw_method=i)
    ax.plot(x, k.pdf(x), label=i)
ax.legend()
ax.set_xlim(0, 150)
```
